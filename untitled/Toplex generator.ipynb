```python
import numpy as np
import hypernetx as hnx
```

```python
def toplex_gen1(ntops=3,start=6,overlap=3):
    '''Returns ntops toplexes of sizes between 
    start + overlap and start + 2*overlap.
    overlap must be less than start '''
    if overlap >=start:
        overlap = start - 1
    toplexes = [list(np.arange(start))]
    othersets = list()
    for i in range(ntops-1):
        j = np.random.randint(overlap+start,overlap+2*start)
        toplexes += [[toplexes[-1][-1*overlap] + k for k in range(j)]]
    for t in toplexes:
        k = np.random.randint(1,len(t))
        for j in range(1,k):
            othersets += [t[:j]]
    return toplexes, toplexes + othersets

tops,alls = toplex_gen1(ntops=3,start=6,overlap=2)

h = hnx.Hypergraph(alls,static=True)
hnx.draw(h)
tops
```
