# HyperNetX
Check out our hypergraph python package at https://github.com/pnnl/HyperNetX. Documentation is available here: https://pnnl.github.io/HyperNetX/build/index.html. 

This demo will show some HNX basics and some real examples of exploratory data analysis using hypergraphs.

```python
# start by building a small hypergraph from some Les Miserables character-scene data

scenes = {
    0: ('FN', 'TH'),
    1: ('TH', 'JV'),
    2: ('BM', 'FN', 'JA'),
    3: ('JV', 'JU', 'CH', 'BM'),
    4: ('JU', 'CH', 'BR', 'CN', 'CC', 'JV', 'BM'),
    5: ('TH', 'GP'),
    6: ('GP', 'MP'),
    7: ('MA', 'GP')
}

LM = hnx.Hypergraph(scenes)
```

```python
plt.figure(figsize=(7, 5))

# we can use HNX to visualize the hypergraph as an "Euler diagram"
hnx.draw(LM, edges_kwargs={'edgecolor':'gray'})
```

Every hypergraph has a "dual" formed by swapping the roles of verties and edges. We can compute the dual easily in HNX.

```python
LM_D = LM.dual()
```

```python
# incidence_dict gives the dictionary of incidences with edges as keys and sets of vertices as values
LM.incidence_dict, LM_D.incidence_dict
```

```python
plt.figure(figsize=(7, 5))

hnx.draw(LM_D, edges_kwargs={'edgecolor':'gray'})
```

Some simple properties and functions to learn more about the hypergraph

```python
# Nodes as an "EntitySet" object (part of the HNX package, underlying hypergraph data structure)
LM.nodes
```

```python
# nodes just as a list of names
list(LM.nodes)
```

```python
# edges just as a list of names, without "list" returns the EntitySet object
list(LM.edges)
```

```python
# (number of nodes, number of edges)
LM.shape
```

```python
# degree is number of edges a node is contained within
LM.degree('FN')
```

```python
# you can see what the edges are using the nodes and membership properties
list(LM.nodes['FN'].memberships)
```

```python
# the size of an edge is the number of nodes within it
LM.size(3)
```

```python
# you can see the nodes in an edge using the edges property
list(LM.edges[3])
```

s-components and s-distance can also be computed easily

```python
# s_connected_components with edges=True returns a generator for the s-edge-connected components
for comp in LM.s_connected_components(s=2, edges=True):
    print(comp)
```

```python
# s_connected_components with edges=False returns a generator for the s-node-connected components
for comp in LM.s_connected_components(s=2, edges=False):
    print(comp)
```

```python
# "distance" computes node distance. Default s=1
LM.distance('MA', 'FN')
```

```python
# "edge_distance" computes edge distance. Default s=1
LM.edge_distance(4, 6)
```

```python
# if two edges aren't s-connected then the distance is infinity
LM.edge_distance(4, 6, s=2)
```

```python
# node and edge s-diameter is longest s-distance. Default s=1
LM.diameter(), LM.edge_diameter()
```

```python
# uncomment to show error when hypergraph is not s-connected
#LM.diameter(s=2)
```

```python
# node_diameters (default s=1) returns 
#    maximum node diameter over all components
#    list of node diameters for each component
#    list of node sets for each component

LM.node_diameters(s=2)
```

```python
# edge_diameters (default s=1) returns 
#    maximum edge diameter over all components
#    list of edge diameters for each component
#    list of edge sets for each component

LM.edge_diameters(s=2)
```

We can also compute s-betweenness and s-harmonic closeness centrality

```python
hnx.s_betweenness_centrality(LM)
```

```python
hnx.s_harmonic_closeness_centrality(LM)
```

```python
hnx.s_betweenness_centrality(LM_D)
```

```python
hnx.s_harmonic_closeness_centrality(LM_D)
```

When a group of nodes are in the same set of edges they can be "collapsed".

```python
LM_collapsed = LM.collapse_nodes()
```

```python
plt.figure(figsize=(7, 5))

hnx.draw(LM_collapsed)
```

```python
%%time
# read in prior data as connected components for s=1, s=2, s=3

datadir = 'C:/Users/hoga886/Documents/_Projects_code_work/Old/_Hypernets/DNS_Data/CHGL/CHGL-DNS-Metrics/'
day = 'FullDay/'
comps_file = 'components-FULLDAY.txt'

comps_dict = {}

with open(datadir+day+comps_file,'r') as compsf:
    in_s = False
    for line in compsf:
        line = line.strip()
        if line[:25] == 'Edge Connected Components':
            s = int(line[-3])
            comps_dict[s] = {}
            in_s = True
        elif line[:27] == 'Vertex Connected Components':
            in_s = False
        elif line[:9] == 'Component' and in_s:
            comp_num = int(line.split('#')[-1][:-1])
            comps_dict[s][comp_num] = {}
        elif in_s:
            line = line.strip()
            domain, IPs = line.split('\t')
            comps_dict[s][comp_num][domain] = IPs.split(',')
```

```python
# for the components with between 5 and 100 edges what are the diameters and shapes

comps_diams = []
for idx in comps_dict[s]:
    if len(comps_dict[s][idx]) < 100 and len(comps_dict[s][idx]) > 5:
        temp = hnx.Hypergraph(comps_dict[s][idx])
        try:
            comps_diams += [(idx, temp.edge_diameter(s=s), temp.shape)]
        except:
            continue
```

```python
# distribution of those diameters

vals, counts = np.unique([c[1] for c in comps_diams], return_counts=True)
plt.scatter(vals, counts)
plt.yscale('log')
plt.xscale('log')
```

## Transcriptomics data
In computational biology measuring expression of gene transcripts is called "transcriptomics". Given a biological sample it's possible to measure the expression level of genes through microarray or high throughput sequencing technologies. Typical experiments, such as the ones we study here, consider a set of samples (e.g., a time course) relative to a control in the form of a “$\log_2$ fold change” computed by taking the mean of $\log_2$ of all experimental replicates and subtracting mean of $\log_2$ of all control replicates for each gene. Positive values mean that the gene is up-regulated compared to the control while negative values indicate that the gene is down-regulated relative to the control.

This data file contains $\log_2$ fold change values (columns ending in "\__b") as well as significance values (columns ending in "\__p") for each of 279 experimental conditions. Each row corresponds to a gene, with the name of the gene as the first element in the row. 

We will create a hypergraph from these transcriptomics data where the vertices represent *conditions* and the hyperedges represent *genes*. A condition vertex will be contained within a gene hyperedge if the z-score (condition-wise) of the $\log_2$ fold change is $\geq 2$. 

```python
tdfb = pickle.load(open('./bigTrans3_anon_b.pikl', 'rb'))
tdfb.head()
```

```python
tdfb.shape
```

```python
# grab a subset of the data so it's easier to work with

tdfb_sample = tdfb[list(tdfb.columns)[0:27]].iloc[0:976]
```

```python
# compute absolute value of z-score over each column

from scipy.stats import zscore

tdfbz = abs(tdfb_sample.apply(zscore))
tdfbz.head()
```

```python
# create hypergraph from dataframe using threshold >=2

H_trans = hnx.Hypergraph.from_dataframe(tdfbz, transpose=True, key = (lambda x : x >= 2.0))
```

```python
H_trans.shape
```

```python
def avg_s_betweenness_centrality(H, mins=1, maxs=1, normalized=True, verbose=False):
    # First compute all s-betweenness centralities for s between mins and maxs (inclusive)
    betc = {}
    for s in range(mins, maxs+1, 1):
        if verbose: # for a large range of s values, or for low s values, a status update can be helpful
            print(s)
        betc[s] = hnx.s_betweenness_centrality(H, s=s, normalized=normalized)
        
    # For each gene in the betc[mins] dictionary we compute its average centrality
    # value across all s values. If that gene does not appear in a dictionary for 
    # a higher s value we say that it has centrality 0 for that s value.
    gene_betc_list = {}
    avg_betcs = {}
    for gene in betc[mins]:
        gene_betc_list[gene] = [betc[s][gene] if gene in betc[s] else 0 for s in betc]
        avg_betcs[gene] = np.mean(gene_betc_list[gene])
    
    return betc, gene_betc_list, avg_betcs


def avg_s_harmonic_closeness_centrality(H, mins=1, maxs=1, verbose=False):
    # First compute all harmonic s-closeness centralities for s between mins and maxs (inclusive)
    cloc = {}
    for s in range(mins, maxs+1, 1):
        if verbose: # for a large range of s values, or for low s values, a status update can be helpful
            print(s)
        cloc[s] = hnx.s_harmonic_closeness_centrality(H, s=s)
        
    # For each gene in the cloc[mins] dictionary we compute its average centrality
    # value across all s values. If that gene does not appear in a dictionary for 
    # a higher s value we say that it has centrality 0 for that s value.
    gene_cloc_list = {}
    avg_clocs = {}
    for gene in cloc[mins]:
        gene_cloc_list[gene] = [cloc[s][gene] if gene in cloc[s] else 0 for s in cloc]
        avg_clocs[gene] = np.mean(gene_cloc_list[gene])
    
    return cloc, gene_cloc_list, avg_clocs
```

```python
betc, gene_betc_list, avg_betcs = avg_s_betweenness_centrality(H_trans, mins=1, maxs=5, verbose=True)
cloc, gene_cloc_list, avg_clocs = avg_s_harmonic_closeness_centrality(H_trans, mins=1, maxs=5, verbose=True)
```

```python
def draw_with_node_colors(H, colordict, cmap_name = 'YlOrRd', kwargs={}, return_pos=False, alpha=1.0):
    import matplotlib.cm as cm
    # normalize the node color dictionary to have values in (0, 1)
    m = min(colordict.values())
    colordict = {k:v-m+0.001 for (k,v) in colordict.items()}
    M = max(colordict.values())
    colordict = {k:v/M for (k,v) in colordict.items()}
    
    # make the list of colors in the correct order
    cmap = cm.get_cmap(cmap_name)
    nodes = list(H.nodes)
    colorlist = [(cmap(colordict[node]) if node in colordict else cmap(0)) for node in nodes]
    
    # use the hnx drawing function
    if return_pos:
        pos = hnx.draw(H, nodes_kwargs={'facecolors': colorlist, 'alpha':alpha}, return_pos=True, **kwargs)
        return pos
    else:
        hnx.draw(H, nodes_kwargs={'facecolors': colorlist, 'alpha':alpha}, **kwargs)
```

```python
# draw the (dual) hypergraph with genes colored by betweenness

draw_with_node_colors(H_trans.dual(), avg_betcs, 
                      kwargs={'with_node_labels':False, 
                              'with_edge_labels':False, 
                              'edges_kwargs':{'edgecolor':'gray'}}, alpha=0.75)
```

```python
# draw the (dual) hypergraph with genes colored by harmonic closeness

draw_with_node_colors(H_trans.dual(), avg_clocs, 
                      kwargs={'with_node_labels':False, 
                              'with_edge_labels':False, 
                              'edges_kwargs':{'edgecolor':'gray'}}, alpha=0.75)
```
