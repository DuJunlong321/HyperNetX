```python

import hypernetx as hnx
import matplotlib.pyplot as plt
```

```python
u,v,w = 'Blue','Orange','Green'
a,b,c,d,e,f,g,h,i,j,l,m,n,o,p = list('ABCDEFGHIJLMNOP')
pos = {b:(0,5),a:(1,1),c:(2,7),d:(3,1),m:(4,3),e:(5,4),l:(4,5),
      g:(7,1), f:(7,7), p:(10,3), h:(9,4), n:(10,6), o:(9,7),
      i:(12,6),j:(13,5)}
```

```python
edict2 = {u:{a,b,c,d,e,l,m},
        v:{l,e,m,f,g,o,n,h,p},
        w:{o,n,h,p,i,j}}
edict1 = {u:{a,b,c,d,e,},
        v:{e,f,g,h},
        w:{h,i,j}}
```

```python
fig,ax = plt.subplots(1,1,figsize=(10,6))
ax.axis('off')
h1 = hnx.Hypergraph(edict1)
hnx.draw(h1,pos=pos,ax=ax,
         edges_kwargs={'linewidth':4},
         edge_labels_kwargs={'fontsize':20},
         label_alpha=.1,
        node_labels_kwargs={'fontsize':18})
```

```python
fig,ax = plt.subplots(1,1,figsize=(10,6))
ax.axis('off')
h2 = hnx.Hypergraph(edict2)
hnx.draw(h2,pos=pos,ax=ax,
         edges_kwargs={'linewidth':4},
         edge_labels_kwargs={'fontsize':20},
         label_alpha=.1,
        node_labels_kwargs={'fontsize':18})
```
