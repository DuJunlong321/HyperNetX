```python
import hypernetx as hnx
import hypernetx.experimental as hnxe
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
import sys
```

```python
## Random hrmat hypergraph will be more sparse by adjusting the nv parameter
nrows,ncols,res,nv = 3,2,6,4
alpha = 0.1

p = alpha*np.random.rand(nrows*ncols)

ndx,edx,m = hnxe.rmat_hypergraph(nrows,ncols,p,res,nv,return_matrix=True)
m = m.astype(bool)*1
```

```python
## s-nearest neighbors
s = 1
snn = defaultdict(list)
for cdx in range(m.shape[1]): 
    if np.sum(m[:,cdx])<s: ## look at each column and see if has size at least s
        print(cdx,'empty')
        snn[cdx] = list()
    else: 
        snn[cdx].append(cdx)
        nzr = [r for r in range(m.shape[0]) if m[r,cdx] > 0]  ## non zero rows to slice out           
        temp = [c for c in range(cdx+1,m.shape[1]) if np.sum(m[nzr,c]) >= s]
        for c in temp:
            snn[c].append(cdx)
        snn[cdx].extend(temp) ## neighbors share at least s nonzero rows with cdx
snn
   
    
```
