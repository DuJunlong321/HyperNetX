# Read in hypergraph from vertex-hyperedge membership list

```python
#read in data as pandas dataframe
df=pd.read_csv('cubedr_IR.csv', sep=',',header=None)
df=df.rename(columns={0: "ip", 1: "mal",2:'weight'}) #rename columns
df=df[df['mal']!='None'] #drop any vertex hyperedge pairs that are "None"
df
```

```python
#restrict data to largest connected component

#process hypergraph as bipartite graph
B = nx.Graph()

# Add nodes with the node attribute "bipartite"
B.add_nodes_from(df['ip'], bipartite=0) #ip
B.add_nodes_from(df['mal'],bipartite=1)#malwares

#add edges
edges=zip(df['ip'].to_list(),df['mal'].to_list(),df['weight'].to_list()) #edges w/scores
B.add_weighted_edges_from(edges)

#restrict to largest connected components
lar=max(nx.connected_component_subgraphs(B), key=len)
df=df[df['ip'].apply(lambda x: x in lar.nodes())]
```

# Form transition matrix from hypergraph and weights

```python
#Create weighted incidence matrix
verMap={y: x for x,y in enumerate(set(df['ip'])) } #integer indices for vertices
edgeMap={y: x for x,y in enumerate(set(df['mal'])) } #integer indices for hyperedges

#generate incidence matrix as coo matrix
R=coo_matrix((list(df['weight']),\
    (list(df['mal'].apply(lambda x: edgeMap[x])),list(df['ip'].apply(lambda x: verMap[x])))),\
             shape=(len(edgeMap), len(verMap))) #vertices as columns, hyperedges as rows
```

```python
#generate hyperedge weight matrix.
#has same nonzero pattern as incidence matrix. 
#uses chitra's suggestion of making hyperedge weight depend on 
#edge-dependent vertex weight standard deviation
edgeScore={i:np.std(R.getrow(i).data)+1 for i in range(R.shape[0]) } #hyperedge weights
vals=[edgeScore[i] for i in R.nonzero()[0]]

#generate incidence matrix
W=coo_matrix((vals,\
    (R.nonzero()[1],R.nonzero()[0])),\
             shape=(R.shape[1], R.shape[0]))
```

```python
#generate diagonal matrices used to normalize probability transition matrix
[rowSums]=R.sum(axis=1).flatten().tolist()
D_E=diags([1/x for x in rowSums])

[rowSums]=W.sum(axis=1).flatten().tolist()
D_V=diags([1/x for x in rowSums])
```

```python
#put all the pieces together to build the probability transition matrix
#this represents the random walk
P=D_V*W*D_E*R
```

# Form Laplacian from transition matrix

```python
#functions to form different Laplacian matrices

#stationary distribution vector is used to form Laplacians
def get_pi(P):
    rho,pi=(sp.sparse.linalg.eigs(np.transpose(P), k=1, return_eigenvectors=True))
    pi=np.real(pi/np.sum(pi))
    return pi.flatten()

#combinatorial laplacian
def comb_lap(P):
    pi=get_pi(P)
    return np.diag(pi)-(np.diag(pi)*P+np.transpose(P)*np.diag(pi))/2

#normalized laplacian
def norm_lap(P):
    pi=get_pi(P)
    return np.identity(P.shape[0])-\
        (np.diag(np.power(pi,1/2))*P*np.diag(np.power(pi,-1/2))+\
        np.diag(np.power(pi,-1/2))*np.transpose(P)*np.diag(np.power(pi,1/2)))/2

#the non-negative part of the normalized Laplacian used in our algorithms
def norm_lap_core(P):
    pi=get_pi(P)
    return (np.diag(np.power(pi,1/2))*P*np.diag(np.power(pi,-1/2))+\
        np.diag(np.power(pi,-1/2))*np.transpose(P)*np.diag(np.power(pi,1/2)))/2
```

```python
#form normalized laplacian core. This is the final input for spectral and NMF clustering!
L=norm_lap_core(P)
```

# Spectral clustering algorithm using Laplacian

```python
#Spectral clustering algorithm. 
k=6 #choose number of clusters

#compute eigenvectors
e,v=sp.sparse.linalg.eigs(norm_lap_core(P), k=k, which='LM', return_eigenvectors=True)
v=np.real(v) #ignore zero complex parts
v=preprocessing.normalize(v, norm='l2',axis=1) #normalize
U = np.array(v)
km = KMeans(init='k-means++', n_clusters=k,random_state=0) #k-means
km.fit(U)
d=km.labels_
```

# Summarizing everything with main function

```python
def specClus(k,verInc, edgeInc,weights=None):
    """
    Input: vertex-hyperedge memberships with optional weights, and number of clusters, k. 
    Vertex-hyperedge memberships and weights must be specified as equal-length lists, e.g. 

    ['a','a','b'],['e','f','c'],[1,7,1]

    which denotes the hypergraph in which vertex 'a' belongs to hyperedge 'e' with weight 1, 
    'a' belongs to 'f' with weight 7, etc.
    Hypergraph must be connected! 
    Weights must be positive, real numbers. 
    If no weights specified, then all weights are 1. 

    Output: Cluster assingments from hypergraph spectral clustering using normalized Laplacian
    based on edge-dependent vertex weighted random walks. 
    """

    if weights==None:
        weights=[1 for i in range(len(verInc))]
        
    #Create weighted incidence matrix
    verMap={y: x for x,y in enumerate(set(verInc)) } #integer indices for vertices
    edgeMap={y: x for x,y in enumerate(set(edgeInc)) } #integer indices for hyperedges

    #generate incidence matrix as coo matrix
    R=coo_matrix((weights,\
        ([edgeMap[x] for x in edgeInc],[verMap[x] for x in verInc])),\
                 shape=(len(edgeMap), len(verMap))) #vertices as columns, hyperedges as rows
    
    #generate hyperedge weight matrix.
    #has same nonzero pattern as incidence matrix. 
    #uses chitra's suggestion of making hyperedge weight depend on 
    #edge-dependent vertex weight standard deviation
    edgeScore={i:np.std(R.getrow(i).data)+1 for i in range(R.shape[0]) } #hyperedge weights
    vals=[edgeScore[i] for i in R.nonzero()[0]]

    #generate incidence matrix
    W=coo_matrix((vals,\
        (R.nonzero()[1],R.nonzero()[0])),\
                 shape=(R.shape[1], R.shape[0]))
    
    #generate diagonal matrices used to normalize probability transition matrix
    [rowSums]=R.sum(axis=1).flatten().tolist()
    D_E=diags([1/x for x in rowSums])

    [rowSums]=W.sum(axis=1).flatten().tolist()
    D_V=diags([1/x for x in rowSums])
    
    #put all the pieces together to build the probability transition matrix
    #this represents the random walk
    P=D_V*W*D_E*R
            
    #stationary distribution vector is used to form Laplacians
    rho,pi=(sp.sparse.linalg.eigs(np.transpose(P), k=1, return_eigenvectors=True))
    pi=np.real(pi/np.sum(pi))
    pi=pi.flatten()
    
    #form normalized Laplacian, L
    L=(np.diag(np.power(pi,1/2))*P*np.diag(np.power(pi,-1/2))+\
        np.diag(np.power(pi,-1/2))*np.transpose(P)*np.diag(np.power(pi,1/2)))/2
    
    #compute eigenvectors of L
    e,v=sp.sparse.linalg.eigs(L, k=k, which='LM', return_eigenvectors=True)
    v=np.real(v) #ignore zero complex parts
    v=preprocessing.normalize(v, norm='l2',axis=1) #normalize
    U = np.array(v)
    km = KMeans(init='k-means++', n_clusters=k,random_state=0) #k-means
    km.fit(U)
    d=km.labels_
    
    #organize cluster assingments in dictionary of form cluster #: ips
    clusters={i:[] for i in range(k)}
    verNames=list(verMap.keys())

    for i in range(len(verMap)): 
        clusters[d[i]].append(verNames[i])
        
    return clusters
```

```python
%%time
#Example usage:
clusters=specClus(6,df['ip'].tolist(),df['mal'].tolist(),df['weight'].tolist())
```
