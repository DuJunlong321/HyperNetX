```python
from untitled_harrypotter import HarryPotter
harry = HarryPotter()
E = harry.entityset
```

```python
mat,rdict,cdict = E.incidence_matrix(index=True)
nodes = E.translate(1,list(rdict.keys()))
edges = E.translate(0,list(cdict.keys()))
mat = mat.todense().astype(bool)*1
h = Hypergraph.from_numpy_array(mat,
                                 node_names=nodes,
                                edge_names = edges)
```
