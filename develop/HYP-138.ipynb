Add the ability to change labels on Entities

```python
%load_ext autoreload
%autoreload 2
import hypernetx as hnx
```

```python
scenes = {
    0: ('FN', 'TH'),
    1: ('TH', 'JV'),
    2: ('BM', 'FN', 'JA'),
    3: ('JV', 'JU', 'CH', 'BM'),
    4: ('JU', 'CH', 'BR', 'CN', 'CC', 'JV', 'BM'),
    5: ('TH', 'GP'),
    6: ('GP', 'MP'),
    7: ('MA', 'GP')
}

H = hnx.Hypergraph(scenes)
```

## With and without node labels

```python
node_labels = {
    'BM': 'Monsieur Bamatabois',
    'BR': 'Brevet',
    'CC': 'Cochepaille',
    'CH': 'Champmathieu',
    'CN': 'Chenildieu',
    'FN': 'Fantine',
    'GP': 'George Pontmercy',
    'JA': 'Javert',
    'JU': 'Judge of Douai',
    'JV': 'Jean Valjean',
    'MA': 'Marius',
    'MP': 'Madame Pontmercy',
    'TH': 'Th√©nardier'
}
```

```python
hnx.draw(H)
```

```python
hnx.draw(H, node_labels=node_labels)
```

```python
# test interaction with collapsing nodes
hnx.draw(H.collapse_nodes(), node_labels=node_labels)
```

## With and without edge labels

```python
edge_labels = {
    i: f'Scene {i}'
    for i in scenes
}

edge_labels
```

```python
hnx.draw(H)
```

```python
hnx.draw(H, edge_labels=edge_labels)
```
