```python
import hypernetx as hnx
```

```python
scenes = {
    0: ('FN', 'TH'),
    1: ('TH', 'JV'),
    2: ('BM', 'FN', 'JA'),
    3: ('JV', 'JU', 'CH', 'BM'),
    4: ('JU', 'CH', 'BR', 'CN', 'CC', 'JV', 'BM'),
    5: ('TH', 'GP'),
    6: ('GP', 'MP'),
    7: ('MA', 'GP')
}

H = hnx.Hypergraph(scenes)
hnx.draw(H)
```

# Now with dictionaries...
For convenience, drawing properties can be passed as dictionaries, i.e., `{node_uuid, edge_uuid} -> property`. For example, we could set the color of a node to red if that node name starts with a J. (We also set the edge colors to black so the colors dont clash.)

```python
colors = {
    v: 'red' if v.startswith('J') else 'black'
    for v in H.nodes
}

hnx.draw(H,
    nodes_kwargs={'color': colors},
    edges_kwargs={'edgecolors': 'black'}
)

# pass in function?
```

# Alternative implementation using defaultdict
The code below is an alternative way to specify the visual properties by setting some of the values directly and using a `defaultdict` to handle the rest.

```python
from collections import defaultdict

colors = defaultdict(lambda: 'black', {
    'JA': 'red',
    'JV': 'red',
    'JU': 'red',
})

hnx.draw(H,
    nodes_kwargs={'color': colors},
    edges_kwargs={'edgecolors': 'black'}
)
```

# An even more concise syntax...
by passing callback functions, we can skip iterating over the nodes and let `hnx` do the work

```python
from collections import defaultdict

hnx.draw(H,
    nodes_kwargs={'color': lambda v: 'red' if v.startswith('J') else 'black'},
    edges_kwargs={'edgecolors': 'black'}
)
```
